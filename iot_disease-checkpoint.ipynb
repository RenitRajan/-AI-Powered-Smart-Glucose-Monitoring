```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from scipy.stats import boxcox
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('iot.csv')
df 
```

```python
from sklearn.preprocessing import LabelEncoder
gender_le = LabelEncoder()
gender_le.fit(['M','F'])
gender_le.classes_
df['Sex'] = gender_le.transform(df['Sex'])
```

```python
ecg_le = LabelEncoder()
ecg_le.fit(df['ChestPainType'])
df['ChestPainType'] = ecg_le.transform(df['ChestPainType'])
ecg_le.classes_
```

```python
ecg_le = LabelEncoder()
ecg_le.fit(df['RestingECG'])
df['RestingECG'] = ecg_le.transform(df['RestingECG'])
ecg_le.classes_
```

```python
exg_le = LabelEncoder()
exg_le.fit(df['ExerciseAngina'])
df['ExerciseAngina'] = exg_le.transform(df['ExerciseAngina'])
exg_le.classes_
```

```python
st_le = LabelEncoder()
st_le.fit(df['ST_Slope'])
df['ST_Slope'] = st_le.transform(df['ST_Slope'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X = df.values[:,:-1]
Y = df.values[:,-1:]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
print("validation accuracy: ",clf.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
print("validation accuracy: ",clf.score(X_test,y_test))
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train, y_train)
print("validation accuracy: ",clf.score(X_test,y_test))
```
